```python
import pandas as pd
import numpy as np

header_data = ['Checking', 'Dur_month', 'Cred_hist', 'Purpose', 'Cred_amt', 'Sav_bonds', 'Employment', 'Installment', \
               'status_sex', 'debtors', 'present_residence', 'property', 'Age', 'other_install', 'Housing', 'existing_credit',\
               'job', 'people_maint', 'Phone', 'Foreign_worker', 'Credit_worthy']

train_data = pd.read_csv("C:/Users/Venkatesh/Dropbox/DS_projects/German_credit/german_data.txt", delim_whitespace = True,\
                         names = header_data)
```

```python
train_data.head()
```

```python
# mapping the columns with object datatype (string to numerical)
for i in train_data.dtypes.index:
    if train_data.dtypes[i] == np.object:
        x = sorted(train_data[i].unique())
        train_data[i] = train_data[i].replace(x, range(len(x)))
```

```python
train_data.head()
```

```python
train_data.corr()
```

```python
#Columns with low correlation with Credit_worthy can be dropped. I am dropping columns with correlation coefficient less
# than 0.15 and greater than -0.15

# So we can consider only for the analysis 'Checking', 'Dur_month', 'Cred_hist', 'Cred_amt', 'Sav_bonds'
credit_analysis = train_data[['Checking', 'Dur_month', 'Cred_hist', 'Cred_amt', 'Sav_bonds', 'Credit_worthy']]
```

```python
credit_analysis.head()
```

```python
credit_analysis.corr()
```

```python
# Now we are going to split the data set in to "Credit_worthy" and "not Credit_worthy" data sets so that we could email
# appropriate offers to 

credit_email = credit_analysis.groupby(['Checking', 'Dur_month', 'Cred_hist', 'Sav_bonds', 'Credit_worthy'])['Cred_amt'].count()
```

```python
credit_email = credit_email.unstack()
```

```python
credit_email = credit_email.fillna(0)
```

```python
credit_email = credit_email.reset_index()
```

```python
credit_email.columns
```

```python
credit_email.columns = ['Checking', 'Dur_month', 'Cred_hist', 'Sav_bonds','CW_yes','CW_no']
```

```python
credit_email.head()
```

```python
#Now we can convert the ordinal numbers into corresponding text so that we can easily understand and create dashboard.
Dict_checking = {0: 'Low', 1:'Medium', 2:'High', 3:'No Acc'}
Dict_Credit = {0:'Outstanding', 1:'Excellent', 2:'Good', 3:'Bad', 4:'Worst'}
Dict_Sav = {0:'Very Low', 1:'Low', 2:'Medium', 3:'High', 4:'No Acc'}

Dict = [Dict_checking, Dict_Credit, Dict_Sav]
```

```python
list = ['Checking','Cred_hist', 'Sav_bonds']
for i in range(3):
    credit_email[list[i]] = credit_email[list[i]].map(Dict[i])
```
